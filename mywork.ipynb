```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
df=pd.read_csv("machine failure.csv")
df.head()
```

```python
x=df.drop('Machine failure',axis=1)
x
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Identify columns with object data type
object_cols = x.select_dtypes(include='object').columns

# Convert object columns to float
for col in object_cols:
    x[col] = x[col].str.extract('(\d+)').astype(float)

# Scale numeric columns using MinMaxScaler
minmax = MinMaxScaler()
train_minmax = minmax.fit_transform(x)

# Create a new DataFrame with scaled values
x = pd.DataFrame(train_minmax, columns=x.columns)
```

```python
y=df['Machine failure']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer object with the desired strategy
imputer = SimpleImputer(strategy='mean')

# Impute missing values in x_train
x_train_imputed = imputer.fit_transform(x_train)

# Fit the RandomForestClassifier on the imputed data
model_rf.fit(x_train_imputed, y_train)
```

```python
# Impute missing values in x_test using the same imputer object
x_test_imputed = imputer.transform(x_test)

# Make predictions on the imputed test data
y_pred = model_rf.predict(x_test_imputed)
```

```python
model_rf.score(x_test_imputed,y_test)
```
